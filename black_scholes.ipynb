```python
from bs import BlackScholes
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parameter Definitions
X = 100
t = 1
r = 0.01
sgma = 0.3
```

```python
S = np.arange(60,140,0.1)
calls = [BlackScholes(s, X, t, r, sgma).call() for s in S]
puts = [BlackScholes(s, X, t, r, sgma).put() for s in S]
plt.plot(S, calls, label='Call Value')
plt.plot(S, puts, label='Put Value')
plt.xlabel("Underlying Stock Price")
plt.ylabel("Option Value")
plt.title("Effect of Stock Price on Option Value (assuming all else is constant)")
plt.legend()
plt.show()
```
